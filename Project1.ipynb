```python
# Calculate the mean, median, and mode of the Total IFC investment as approved by Board (Million USD)
mean_investment = project['Total IFC investment as approved by Board(Million - USD)'].mean()
median_investment = project['Total IFC investment as approved by Board(Million - USD)'].median()
mode_investment = project['Total IFC investment as approved by Board(Million - USD)'].mode()

print("Mean IFC Investment:", mean_investment)
print("Median IFC Investment:", median_investment)
print("Mode IFC Investment:", mode_investment)
```

```python
#Recalculating mean, median and mode without pandas

import csv
from collections import defaultdict
```

```python
investments = []
with open('World Bank_IFC Investment Services.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        # Convert the value to float if it exists and append to the list
        try:
            investment = float(row['Total IFC investment as approved by Board(Million - USD)'])
            investments.append(investment)
        except ValueError:
            # Skip rows where the investment value is not a number (e.g., empty or NaN values)
            continue
```

```python
#Calculating mean
mean_investment = sum(investments) / len(investments) if investments else 0

sorted_investments = sorted(investments)
mid = len(sorted_investments) // 2
if len(sorted_investments) % 2 == 0:
    median_investment = (sorted_investments[mid - 1] + sorted_investments[mid]) / 2
else:
    median_investment = sorted_investments[mid]

# Calculating mode
count_dict = defaultdict(int)
for investment in investments:
    count_dict[investment] += 1

# Find the investment with the highest count
max_count = max(count_dict.values(), default=0)
mode_investment = [key for key, count in count_dict.items() if count == max_count]

print("Mean IFC Investment:", mean_investment)
print("Median IFC Investment:", median_investment)
print("Mode IFC Investment:", mode_investment)
```

```python
import csv

# Load and bin data
investments = []
with open('World Bank_IFC Investment Services.csv', 'r') as file:
    investments = [float(row['Total IFC investment as approved by Board(Million - USD)']) for row in csv.DictReader(file) if row['Total IFC investment as approved by Board(Million - USD)']]

# Generate histogram data
bins = range(0, int(max(investments)) + 50, 50)
histogram = {f"{b}-{b+49}": sum(b <= x < b + 50 for x in investments) for b in bins}

# Set max bar length and scale bars accordingly
max_bar_length = 1000
max_count = max(histogram.values())
for label, count in histogram.items():
    bar_length = int((count / max_count) * max_bar_length) if max_count else 0
    print(f"{label:>7}: {'*' * bar_length} ({count})")
```
